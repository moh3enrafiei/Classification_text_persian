```python
with open('../Datas/text/stopwords.txt', encoding='utf-8') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [line.replace('\n','') for line in stopwords]
stopwords
```

```python
stopwords = [line.replace('\u200c','') for line in stopwords]
stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
nltk_stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(nltk_stopwords)
len(stopwords)
```

```python
import hazm
stemmer = hazm.Stemmer()
```

```python
stemmer.stem('سبزیجات')
```

```python
from hazm import word_tokenize
```

```python
dataset = pd.DataFrame(columns=('title_body', 'category'))
for index, row in data.iterrows():
    title_body = row['Title'] + ' ' + row['Body']
    title_body_tokenized = word_tokenize(title_body)
    title_body_tokenized_filtered = [w for w in title_body_tokenized if not w in stopwords]
    title_body_tokenized_filtered_stemmed = [stemmer.stem(w) for w in title_body_tokenized_filtered]
    dataset.loc[index] = {
        'title_body': ' '.join(title_body_tokenized_filtered_stemmed),
        'category': row['Category2'].replace('\n', '')
    }
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer=TfidfVectorizer()
tfidf_vectorizer.fit(dataset['title_body'])
```

```python
X=tfidf_vectorizer.transform(dataset['title_body'])
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
```

```python
y=la.fit_transform(dataset['category'])
```

```python
y.shape
```

```python
dataset['category'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn import svm
```

```python
svm=svm.SVC()
svm.fit(x_train,y_train)
```

```python
svm.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
y_pred=svm.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
